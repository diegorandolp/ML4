# Process audio files

```python
# collect all the files in the directory and store them
sound_files = os.listdir(r'.\train\train')
sound_files_path = [os.path.join(r'.\train\train', file) for file in sound_files]
    
```

```python
print(sound_files_path[0])
```

```python
# process the files to tensors
waveforms = []
srs = []
for file in sound_files_path:
    waveform, sr = torchaudio.load(file)
    waveforms.append(waveform)
    srs.append(sr)
```
